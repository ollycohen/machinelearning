```
import numpy as np, matplotlib.pyplot as plt
```

```
#Reference: https://github.com/ppaquay/Learning-from-Data-Solutions/blob/master/Problems_Chap2.pdf

# 2.24 b Describe an experiment that you can run to determine
# gbar(x), Eout, bias, and var.

# Plot the target function f(x) = x^2
# Run an experiment n times:
#   Choose {x1, x2} on the interval [-1,1],
#   Plot the points (x1, x1^2), (x2, x2^2).
#   Then the hypothesis g(x) is the line (of form ax + b) that passes through
#   the two points. Save the hypothesis to a set of all hypotheses
# After the experiment completes, calculate gbar(x) is the average of the 
# set of g(x)


#Plot x^2
inp = np.linspace(-1,1,100)
outp = inp**2
plt.xlim(-1,1)

N = 10000

aSet = np.ones([N,1])
bSet = np.ones([N,1])
#run the experiment N times
for i in range(N):
  # Gather samples
  sign = np.random.choice([-1,1],size=2)
  dist = np.random.rand(1,2)
  x = sign*dist
  y = x**2
  plt.plot(x[0], y[0], color='green', linewidth=.05)
  gx = np.polyfit(x[0], y[0], 1)
  aSet[i] = gx[0]
  bSet[i] = gx[1]

plt.plot(inp, outp, label="f(x)", color='red')


abar = np.mean(aSet)
bbar = np.mean(bSet)
plt.plot(inp, abar*inp + bbar, label="$\overline{g}$(x)", color='purple')

#produce testing set
t_x = np.random.rand(N)
sign = np.random.choice([-1,1], N)
t_y = (t_x * sign)**2
f_mean = np.mean(t_y)

#Formatting the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Learning Target Function f(x)=$x^2$')
plt.legend()

#Calculate results of the experiment 
eout = (aSet* t_x + bSet - t_y)**2
#eout2 = np.mean(t_x**4) - 2*aSet*np.mean(t_x**3) + (aSet**2 - 2*bSet) * np.mean(t_x**2) + 2 * aSet *bSet* np.mean(t_x) + bSet**2
#eout2 = np.mean(eout2)
eout = np.mean(eout)

bias = (abar * t_x + bbar - t_y)**2
bias = np.mean(bias)
print("Bias is {}".format(bias))

var = ((aSet*t_x + bSet) - (abar*t_x + bbar))**2
# var_alt = np.mean((aSet[0] - abar)**2)* t_x**2 + 2*np.mean((aSet[0]-abar)*(bSet[0]-bbar)) *t_x + np.mean((bSet[0]-bbar)**2)
# var_alt = np.mean(var_alt)
var = np.mean(var)
print("Var is {}".format(var))
print("Eout is {} = Bias + Variance {}".format(eout, bias+var))



```
